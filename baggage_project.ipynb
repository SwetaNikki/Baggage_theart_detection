```python
#cropping the object from image
def crop_object_from_image(image):
    # Load image, convert to grayscale, Gaussian blur, Otsu's threshold
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3,3), 0)
    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Obtain bounding rectangle and extract ROI
    x,y,w,h = cv2.boundingRect(thresh)
    cv2.rectangle(image, (x, y), (x + w, y + h), (36,255,12), 2)
    ROI = original[y:y+h, x:x+w]

    # Add alpha channel
    b,g,r = cv2.split(ROI)
    alpha = np.ones(b.shape, dtype=b.dtype) * 50
    ROI = cv2.merge([b,g,r,alpha])

    cv2.imshow('thresh', thresh)
    cv2.imshow('image', image)
    cv2.imshow('ROI', ROI)
    cv2.imwrite('Image_crop.jpg', ROI)
    cv2.waitKey()
    

    
    
#resizing the object_image
def resize_image(image):
    #Create an Image Object from an Image
    im = Image.open(image)
    #im = im.convert('RGB')

    #Make the new image half the width and half the height of the original image
    resized_im = im.resize((round(im.size[0]*0.5), round(im.size[1]*0.5)))

    #Display the resized imaged
    resized_im.show()

    #Save the cropped image
    resized_im.save('resized-image.jpg')


    
    
#rotating the object_image to 45 degree   
def rotate_image(image):
    # creating a image object (main image) 
    im2 = Image.open(image) 
    #im2 = im2.convert('RGB')

    # rotating a image 45 deg counter clockwise 
    im2 = im2.rotate(45, PIL.Image.NEAREST, expand = 0) 
    
    # to show specified image 
    im2.save('rotate_img.jpg') 
    im2.show('rotate_img.jpg')
    


    
    
    
#converting the object_image white background to transparent   
def converting_white_background_to_tranparent(image):
    img = Image.open(image)
    
    #to have the image color as it is
    img = img.convert("RGBA")
    img.putalpha(128) #making the image little transparent to have a real look with the baggage
    datas = img.getdata()
    
    newData = []
    
    #for loop to eliminate the white range color
    for item in datas:
        # if any pixel color lies in this white color range, then append the list to (255,255,255)
        if ((item[0] == 255 and item[1] == 255 and item[2] == 255) or (item[0] == 245 and item[1] == 245 and item[2] == 245) or
            (item[0] == 248 and item[1] == 248 and item[2] == 255) or (item[0] == 255 and item[1] == 254 and item[2] == 250) or
            (item[0] == 255 and item[1] == 250 and item[2] == 250) or (item[0] == 255 and item[1] == 255 and item[2] == 240) or
           (item[0] == 255 and item[1] == 250 and item[2] == 240) or (item[0] == 255 and item[1] == 245 and item[2] == 238) or
            (item[0] == 255 and item[1] == 248 and item[2] == 220) or (item[0] == 253 and item[1] == 245 and item[2] == 230) or
            (item[0] == 255 and item[1] == 253 and item[2] == 208) or (item[0] == 245 and item[1] == 245 and item[2] == 220) or
           (item[0] == 241 and item[1] == 233 and item[2] == 210) or (item[0] == 250 and item[1] == 235 and item[2] == 215) or
            (item[0] == 247 and item[1] == 231 and item[2] == 206) or (item[0] == 240 and item[1] == 234 and item[2] == 214) or
            (item[0] == 239 and item[1] == 223 and item[2] == 187) or (item[0] == 227 and item[1] == 218 and item[2] == 201) or
           (item[0] == 243 and item[1] == 229 and item[2] == 171) or (item[0] == 238 and item[1] == 220 and item[2] == 130) or
            (item[0] == 237 and item[1] == 234 and item[2] == 224)):
            newData.append((255, 255, 255, 0))
        else:
            newData.append(item)
            
    img.putdata(newData)
    
    img.save("white_tranparent.png", "PNG")
    img.show()
    #print("successful")
    



    
    
    
# getting boundary coordinates of a baggage, so that the object which we will paste will remain inside the boundary
def Object_inside_Baggage(image):
    image = cv2.imread(image)
    #original = image.copy()
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3,3), 0)
    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Obtain bounding rectangle and extract ROI
    x,y,w,h = cv2.boundingRect(thresh)
    start = (x, y)
    end = (x + w, y + h)
    return start, end

        

    
    
#pasting the object_image on baggage        
def pasting_object_on_image(baggage_image, object_image):
    filename = object_image
    ironman = Image.open(filename, 'r')
    filename1 = baggage_image
    bg = Image.open(filename1, 'r')
    #text_img = Image.new('RGBA', (600,320), (0, 0, 0, 0))
    text_img = Image.new('RGBA', (340,400), (0,0,0,0))
    
    #taking the starting and ending coordinates of baggage image
    start, end = Object_inside_Baggage(filename1)
    
    #setting the location of object image inside baggae boundary
    #can change if required
    x = 50
    y = 160
    #condition for if the object_image is inside the boundary of the baggage
    if (x in range(start[0], end[0])) and (y in range(start[1], end[1])):
        text_img.paste(bg, (0,0))
        text_img.paste(ironman, (x,y), mask=ironman)
        text_img.save("output.png", format="png")
        text_img.show()
    else:
        #if its not, change the coordinates
        print("OBJECT IS GOING OUT OF BOUNDARY OF BAGGAGE...PLEASE CHANGE THE COORDINATES")

        
```
